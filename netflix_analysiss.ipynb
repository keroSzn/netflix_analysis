```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv', encoding='iso-8859-1')
```

```python
df.head()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20',
                     'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'])
df
```

```python
df.isna().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)

df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
```

```python
df['type'].value_counts().plot(kind='pie',figsize=(12,8),title='Distribution of Content Types',autopct='%1.2f%%',ylabel='',shadow=bool)
```

```python
df['release_year'].value_counts().sort_index().plot(kind='bar',figsize=(12,8),title='Content Release Trend Over the Years',xlabel='Release Years',ylabel='Number of Titles',color='brown')
```

```python
df['country'].value_counts().head(10).plot(kind='bar',figsize=(12,8),xlabel='Country',ylabel='Number of Titles',color='cyan',title='Top 10 Countries by Content Production')
```

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='bar',figsize=(12,8),title='Top 10 Genres',xlabel='Genre',color='pink')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.show()
```

```python
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']

plt.figure(figsize=(12, 6))
movies['duration'] = movies['duration'].str.replace(' min', '').astype(int)
sns.histplot(movies['duration'], bins=30, kde=True, legend=False)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()

plt.figure(figsize=(12, 6))
tv_shows['duration'] = tv_shows['duration'].str.replace(' Seasons', '').str.replace(' Season', '').astype(int)
sns.histplot(tv_shows['duration'], bins=10)
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of TV Shows')
plt.show()
```

```python
df_filtered = df[~df['director'].isin(['Unknown', '']) & df['director'].notna()]
df_filtered['director'].value_counts().head(10).plot(kind='bar',figsize=(12,8),title='Top 10 Directors',xlabel='Directors',color='orange')
```

```python
df_filtered = df[~df['cast'].isin(['Unknown', '']) & df['cast'].notna()]
df_filtered['cast'].value_counts().head(10).plot(kind='bar',figsize=(12,8),title='Top 10 Cast Members',xlabel='Cast Members',color='green')
```
